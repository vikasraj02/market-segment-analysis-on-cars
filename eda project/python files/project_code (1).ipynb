```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import re
from bs4 import BeautifulSoup 
```

```python
brand_all = ["Tata","Skoda","Mahindra","Honda","Toyota","Volkswagen","Kia","Jeep","ISUZU","Jaguar","Volvo","Datsun","BMW",
             "Lamborghini","MG-Motor","Nissan","Volvo","Lexus","Porsche","Ferrari","Bentley","mclaren","Bajaj","Maserati",
             "Audi"] # this are all the brand names of the car
brands =[]
models = []
for i in brand_all:    
    page = requests.get("https://www.zigwheels.com/newcars/"+i) #here i means brand name we are itirating brand name to i
    url = BeautifulSoup(page.text)
    b = url.find_all("span",class_="lnk-hvr fnt-16 b block of-hid h-height")
    #brand
    for i in b:
        a = i.text
        brands.append(a.split()[0])#here we are spliting the brand name and model name and appending only brand name to the list
    #model = []
    for i in b:
        a = i.text
        c = a.split()[1:]#here we are spliting the brand name and model name 
        d = " ".join(c)
        models.append(d) #appending only model name
```

# main

```python
brand1 = []
model1 = []
mileage = []
GT = []
variant_type = []
price = []
model1 = []
cc= []
Boot_Space = []
for brand,model in zip(brands,models):
    page = requests.get(f"https://www.zigwheels.com/newcars/{brand}/{model}")
    url = BeautifulSoup(page.text)
    #variant
    for i in url.find_all("span",class_ = "modelName"):
        for j in url.find_all("table",class_ = "zw-bikeover-table ns-spec"):
            a2 = j.text
            b2 = re.findall(r"(\d+)\scc",a2) #regular expression check for cc
            c2 = re.findall(r"(\d+)\sLiters",a2)#regular expression check for boot space
            #cc
            if len(b2)>0:#In some places there are empty list of cc so we are checking wether the length of b is >1 
                cc.append(b2[0])#if the length b is >1 then only append
            else:
                cc.append(np.nan)#if the length b is  < 1 then append Nan
            #Boot space
            if len(c2)>0:
                Boot_Space.append(c2[0])#if the length of c is >1 then only append
            else:
                Boot_Space.append(np.nan)#if the length of c is >1 then only append
        a = i.text 
        variant_type.append(a)
        model1.append(model) #Here we are appending model because to model row should match with variant
        brand1.append(brand)#Here we are appending brand because to brand row should match with variant
    
    #geartype
    for i in url.find_all("div",class_ = "fnt-12 clr-bl o-7 clr"):
        a = i.text
        split = a.split() #Here we are spliting the table and taking only geare type
        GT.append(split[-1])# we are appending gear type
    #price
    for i in url.find_all("span",class_ = "b"):
        a = i.text
        b = re.findall(r"Rs\W\s(\d+\W\d+\s\w+)",a) #Rs. 8.41 Lakh
        if len(variant_type) == len(price): #here we are comparing the length of the price with the variant type 
            break # if the len of the price and the variant type is same then we are brekng the loop
        else:
            if len(b)>0: # If the len of the b > 0 then this condition will be satisfy
                price.append(b[0]) #if the condtion got satisfied then it will append to price list
    
```

```python
l = [brand1,model1,GT,variant_type,price,cc,Boot_Space]
for i in l:
    print(len(i))
```

```python
# we are creating data frame
main_df = pd.DataFrame({"brand":brand1,"model":model1,"variant_type":variant_type,"CC":cc,"gear_type":GT,"Boot_Space":Boot_Space,"price":price})
```

# brand

```python
brand_all = ["Maruti-Suzuki","Land-Rover","Rolls-Royce","Force-Motors","MINI","Aston-Martin",""]
brands2 = []
models2 = []
brands22 = []
for i in brand_all:
    page = requests.get("https://www.zigwheels.com/newcars/"+i)
    url = BeautifulSoup(page.text)
    b = url.find_all("span",class_="lnk-hvr fnt-16 b block of-hid h-height")
    #brand =[]
    for i in b:
        a = i.text
        c = a.split()[0:2]
        d = " ".join(c)
        brands2.append(d)
    #model = []
    for i in b:
        a = i.text
        c = a.split()[2:]
        d = " ".join(c)
        models2.append(d)
    #brand 
    for i in range(len(brands2)):
        a = re.sub(" ","-",brands2[i])
        brands22.append(a)
        
```

```python
brand2 = []
model2 = []
mileage2 = []
GT2 = []
variant_type2 = []
price2 = []
cc2= []
Boot_Space2 = []
for brand,model in zip(brands22,models2):
    page = requests.get(f"https://www.zigwheels.com/newcars/{brand}/{model}")
    url = BeautifulSoup(page.text)
    #variant
    for i in url.find_all("span",class_ = "modelName"):
        for j in url.find_all("table",class_ = "zw-bikeover-table ns-spec"):
            a2 = j.text
            b2 = re.findall(r"(\d+)\scc",a2)
            c2 = re.findall(r"(\d+)\sLiters",a2)
            #cc
            if len(b2)>0:
                cc2.append(b2[0])
            else:
                cc2.append(np.nan)
            #Boot space
            if len(c2)>0:
                Boot_Space2.append(c2[0])
            else:
                Boot_Space2.append(np.nan)
        a = i.text
        variant_type2.append(a)
        model2.append(model)
        brand2.append(brand)
    #geartype
    for i in url.find_all("div",class_ = "fnt-12 clr-bl o-7 clr"):
        a = i.text
        split = a.split()
        GT2.append(split[-1])
    #price
    for i in url.find_all("span",class_ = "b"):
        a = i.text
        b = re.findall(r"Rs\W\s(\d+\W\d+\s\w+)",a) #Rs. 8.41 Lakh
        if len(variant_type2) == len(price2):
            break
        else:
            if len(b)>0:
                price2.append(b[0])
```

```python
l = [brand2,model2,GT2,variant_type2,price2,cc2,Boot_Space2]
for i in l:
    print(len(i))
```

```python
brand_df = pd.DataFrame({"brand":brand2,"model":model2,"variant_type":variant_type2,"CC":cc2,"gear_type":GT2,"Boot_Space":Boot_Space2,"price":price2})
```

```python
benz_df = pd.DataFrame({"brand":brand3,"model":model3,"variant_type":variant_type3,"CC":cc3,"gear_type":GT3,"Boot_Space":Boot_Space3,"price":price3})
```

```python
final_df = pd.concat([main_df,brand_df,benz_df]) # we are concating the three Data frame into one data frame
```
